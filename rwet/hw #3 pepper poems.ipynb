```python
pepper1 = random.choice(peppers)
letter1 = pepper1[0].lower()

# pepper1 = [x for x in peppers if x[0].lower() == letter1]
firstname1 = [x for x in firstname if x[0].lower() == letter1]
lastname1 = [x for x in lastname if x[0].lower() == letter1]
verb1 = [x for x in verb if x[0] == letter1]
unit1 = [x for x in unit if x[0] == letter1]
adj1 = [x for x in adj if x[0] == letter1]

print(pepper1)
print("")
print(firstname1)
print(lastname1)
print(verb1)
print(unit1)
print(adj1)
```
